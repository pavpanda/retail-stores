```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("retail_data.csv")
df
```

```python
def plot_company(df, company):
    plt.style.use("ggplot")
    x = range(2012, 2020)
    y = np.array(df.loc[df["Company"] == company].iloc[:, -(2019-2012+1):]).flatten()
    plt.plot(x, y)
    plt.title("Revenue of " + company + " vs Time")
    plt.ylabel("Revenue (billion USD)")
    plt.xlabel("Time (year)")
    plt.savefig(company + '.png', bbox_inches="tight")
```

```python
companies = ["Walmart","Target","Amazon"]
for company in companies:
    plot_company(df, company)
```

```python
plot_company(df, "Walmart")
```

```python
plot_company(df, "Amazon")
```

```python
plot_company(df, "Target")
```
